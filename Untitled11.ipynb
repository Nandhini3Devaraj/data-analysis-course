```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
(dataset_train, dataset_test),info = tfds.load(
    'tf_flowers',
    split=["train[:80%]", "train[:80%]"],
    as_supervised=True,
    with_info=True,
)
```

```python
class_names = info.features['label'].names
print("Classes:",class_names)
#preprocessing
def preprocess(image,label):
  image=tf.image.resize(image,(150,150))
  image=image/255.0
  return image,label
```

```python
#apply preprocessing
dataset_train=dataset_train.map(preprocess).batch(32)
dataset_test=dataset_test.map(preprocess).batch(32)
```

```python
#build model
model=models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(512,activation='relu'),
    layers.Dense(5,activation='softmax')
])

```

```python
#compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#train model
history=model.fit(dataset_train,epochs=10)
#evaluate model
loss,acc=model.evaluate(dataset_test)
print("Test Accuracy:",acc)
for images,labels in dataset_test.take(1):
  predictions=model.predict(images)
  index=8
  plt.imshow(images[index])
  plt.axis("off")
```
