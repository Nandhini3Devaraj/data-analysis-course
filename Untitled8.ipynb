```python
import zipfile
import os

zip_file_path = '/content/gold price.zip'
output_dir = '/content/gold_data/'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(output_dir)

print(f"Unzipped '{zip_file_path}' to '{output_dir}'")

# List the contents of the output directory to show what was extracted
print("Contents of the unzipped directory:")
for root, dirs, files in os.walk(output_dir):
    for name in files:
        print(os.path.join(root, name))
    for name in dirs:
        print(os.path.join(root, name))
```

```python
gold_data = pd.read_csv('/content/gold_data/FINAL_USO.csv')
gold_data.head()
```

```python
print(gold_data.shape)
gold_data.info()
```

```python
gold_data.describe()
```

```python
print('Missing values in gold_data before conversion:\n', gold_data.isnull().sum())
```

```python
gold_data['Date'] = pd.to_datetime(gold_data['Date'], errors='coerce')
gold_data.info()
```

```python
y = gold_data['Close']
x = gold_data[['Open', 'High', 'Low']]

print("First five rows of feature set x:")
print(x.head())

print("\nFirst five values of target variable y:")
print(y.head())
```

```python
split_index = int(len(gold_data) * 0.8)

x_train = x.iloc[:split_index]
x_test = x.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
from sklearn import metrics

# 1. Make predictions on the training data
y_train_pred = model.predict(x_train)

# 2. Calculate R-squared error, MAE, and MSE for training data
r2_train = metrics.r2_score(y_train, y_train_pred)
mae_train = metrics.mean_absolute_error(y_train, y_train_pred)
mse_train = metrics.mean_squared_error(y_train, y_train_pred)

print('Training data metrics:')
print(f'R-squared error: {r2_train:.4f}')
print(f'Mean Absolute Error (MAE): {mae_train:.4f}')
print(f'Mean Squared Error (MSE): {mse_train:.4f}')
```

```python
import sklearn.metrics as metrics

# 3. Make predictions on the test data
y_test_pred = model.predict(x_test)

# 4. Calculate R-squared error, MAE, and MSE for test data
r2_test = metrics.r2_score(y_test, y_test_pred)
mae_test = metrics.mean_absolute_error(y_test, y_test_pred)
mse_test = metrics.mean_squared_error(y_test, y_test_pred)

print('Test data metrics:')
print(f'R-squared error: {r2_test:.4f}')
print(f'Mean Absolute Error (MAE): {mae_test:.4f}')
print(f'Mean Squared Error (MSE): {mse_test:.4f}')
```

```python
import matplotlib.pyplot as plt

# 1. Create a scatter plot of Actual Prices vs Predicted Prices
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs. Predicted Prices')
plt.grid(True)
plt.show()

# 2. Create a time-series plot of Actual vs Predicted Prices over time
plt.figure(figsize=(14, 7))
# Ensure the dates correspond to the test set
test_dates = gold_data['Date'].iloc[split_index:]
plt.plot(test_dates, y_test, label='Actual Price', color='blue')
plt.plot(test_dates, y_test_pred, label='Predicted Price', color='red', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Gold Price')
plt.title('Actual vs. Predicted Gold Prices Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
